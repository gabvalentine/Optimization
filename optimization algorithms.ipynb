## Concepto general

Sea una función de una variable independiente $x\in\mathbb{R}$:

\begin{equation}
y=f\left(x\right)
\end{equation}

Supóngase el problema de encontrar el punto $x_{\max}$ o punto óptimo $x_{\textrm{opt}}$el cual haga que la función en tal punto $f\left(x_{\max}\right)$ tome su valor máximo, de modo que:

\begin{equation}
\forall x,\:f\left(x\right)\leq f\left(x_{\max}\right).
\end{equation}

A tal valor $x_{\max}$ se le denomina el **óptimo global** de la función a optimizar, usualmente referida como **función de
costo**, **función objetivo**, o en el aprendizaje automático,
**función de pérdida**. Los **algoritmos de optimización** tienen entonces por objetivo encontrar tal punto óptimo. La siguiente figura
 ilustra los conceptos de óptimo global y local.

![](https://drive.google.com/uc?export=view&id=1KILPQdFA3PRB2r49CnD5oBusczLSmXqL)

La derivada y segunda derivada pueden ser usados como criterio
de optimalidad:

\begin{equation}
f'\left(x\right)=0\qquad f''\left(x\right)>0
\end{equation}

Muchas veces, la evaluación de la función $f\left(x\right)$ es computacionalmente muy costosa, por lo que muchos algoritmos se conforman con buscar un **óptimo local**. Un óptimo local, es un valor óptimo en un vecindario de la función que va desde el punto $x_{a}$ al punto $x_{b}$, lo cual se expresa como: 

\begin{equation}
x\in\left[x_{a},x_{b}\right],\:f\left(x\right)\leq f\left(x_{\max}\right).
\end{equation}

El encontrar un óptimo local es computacionalmente más viable en muchos casos, y la solución encontrada suficiente. A continuación se detallan distintas variantes del problema de optimización, según características de la función a optimizar.

## Optimización convexa y no convexa

Una función convexa cumple que $\forall x,y\in\mathbb{R}$ y $\forall t\in[0,\text{1}]$:

\begin{equation}
f\left(tx+(1-t)y\right)<tf(x)+(1-t)f(y)
\end{equation}

Donde $tx+(1-t)y$ corresponde a todos los puntos en el intervalo $[x,y]$, por lo que la evaluación en la función convexa de tales
puntos $f\left(tx+(1-t)y\right)$ es menor a los puntos en la recta dada entre $f\left(x\right)$ y $f\left(y\right)$, con ecuación $tf(x)+(1-t)f(y)$, como lo ilustra la Figura:

![](https://drive.google.com/uc?export=view&id=1G_jb0cis_AS2qkIMjv5vGOgcA4XiJTBZ)

El negativo $-f$ de una función convexa $f$ es una función cóncava. Observe como el punto óptimo (mínimo en el caso de la función convexa, máximo en la función cóncava) es el óptimo global, por lo que en una función convexa sólo existe un mínimo global. Los algoritmos diseñados para encontrar el punto óptimo en una función tienen por objetivo encontrar un único mínimo global. 

Las funciones no convexas, no cumplen la condición $f\left(tx+(1-t)y\right)<tf(x)+(1-t)f(y)$ con múltiples *valles* y *montañas*, correspondientes a múltiples mínimos y máximos locales, como se ilustra en la Figura:


![](https://drive.google.com/uc?export=view&id=1OUmzrkrCt5Okz8K9wXPI_GWJTVtQEzP7)

Los problemas más frecuentes en aplicaciones cotidianas forman funciones no convexas con múltiples óptimos locales, por lo que los algoritmos diseñados para la optimización de funciones no convexas tienen por objetivo buscar un punto óptimo local lo suficientemente  *satisfactorio*.

Los algoritmos de optimización como el enjambre de partículas, el calentamiento simulado, los algoritmos genéticos, el Newton-Raphson
o el descenso de gradiente son todos algoritmos diseñados para encontrar mínimos locales en funciones no convexas, donde incluso prescinden de una expresión analítica de la función.

## Optimización de funciones diferenciables

La primera derivada en funciones de una dimensión $f\left(x\right)$, denotada como $\frac{df}{dx}$ o $f'\left(x\right)$, y su extensión conceptual en funciones multivariable $f\left(\overrightarrow{x}\right)$ con $\overrightarrow{x}\in\mathbb{R}^{n}$, el vector gradiente $\nabla_{\overrightarrow{x}}f$ indica la dirección hacia donde crece la función, con $\overrightarrow{x}=\left[x_{1},x_{2},\ldots,x_{n}\right]^{T}$,
y el vector gradiente dado más explicitamente por $\nabla_{\overrightarrow{x}}f=\left[\frac{df}{dx_{1}},\frac{df}{dx_{2}},\ldots,\frac{df}{dx_{n}}\right]^{T}$.

La segunda derivada de una función en una variable se expresa como $\frac{df}{dx^{2}}$ o $f''\left(x\right)$ describe la curvatura
o concavidad de la función, y los puntos de inflexión o cortes en cero de la función $f$, como lo ilustra la Figura:

![](https://drive.google.com/uc?export=view&id=13DpBhv1OR0gsEb3umFQxJBVld60j87SV)


De este modo, si $f''\left(x\right)>0$ la función es convexa, y si $f''\left(x\right)<0$ es cóncava.

Para el caso de funciones multivariable $f\left(\overrightarrow{x}\right)$ con $\overrightarrow{x}\in\mathbb{R}^{n}$, la extensión del concepto de la segunda derivada en un espacio de dimensión $n$ se define en la matriz Hessiana (matriz cuadrada de todas las segundas derivadas parciales)  $H_{\overrightarrow{x}}f$: 

\begin{equation}
H_{\overrightarrow{x}}f=\begin{bmatrix}\frac{df}{dx_{1}^{2}} & \frac{df}{dx_{1}dx_{2}} & \ldots & \frac{df}{dx_{1}dx_{n}}\\
\frac{df}{dx_{2}dx_{1}} & \frac{df}{dx_{2}^{2}} & \ldots & \frac{df}{dx_{2}dx_{n}}\\
\vdots & \vdots & \ddots & \vdots\\
\frac{df}{dx_{n}dx_{1}} & \ldots & \ldots & \frac{df}{dx_{n}^{2}}
\end{bmatrix}
\end{equation}

La segunda derivada es entonces más costosa de computar en general. Los métodos basados en la primera derivada como el **descenso
de gradiente**, y en la segunda derivada como el algoritmo de **Newton-Raphson** necesitan entonces de una derivada computable en tiempo reazonable, ya sea analítica o numéricamente. Los métodos basados en la segunda derivada necesitan de menos iteraciones, pues la **información de la curvatura representada en la matriz Hessiana da una mayor exactitud de la localización del óptimo**, sin embargo, es necesario mayor cómputo por iteración, y muchas veces el cálculo de la segunda derivada en expresiones analíticas no es tratable. Ambos algoritmos son estocásticos por cuanto su convergencia depende del punto inicial $x_{0}$, elegido usualmente al azar.

## Optimización convexa: Método de las interpolaciones cuadráticas sucesivas

El algoritmo de la bisección cuadrática es un algoritmo simple en el cual se tiene por objetivo buscar por ejemplo el valor mínimo de
una función convexa $f\left(x\right)$, en el intervalo $\left[a_{0},b_{0}\right]$. Para ello se biseca tal intervalo por la mitad, haciendo: 

\begin{equation}
c_{0}=a_{0}+\frac{b_{0}-a_{0}}{2}
\end{equation}

y se calculan los parámetros de la función cuadrática $w_{0}x^{2}+y_{0}x+z_{0}$ que pase por los tres puntos $f\left(a_{0}\right),f\left(b_{0}\right),f\left(c_{0}\right)$, de forma que debe resolverse el sistema de ecuaciones: 

\begin{equation}
\begin{array}{c}
w_{0}a_{0}^{2}+y_{0}a_{0}+z_{0}=f\left(a_{0}\right)\\
w_{0}b_{0}^{2}+y_{0}b_{0}+z_{0}=f\left(b_{0}\right)\\
w_{0}c_{0}^{2}+y_{0}c_{0}+z_{0}=f\left(c_{0}\right)
\end{array}
\end{equation}

Luego de resolverse tal sistema y conocer los valores de los parámetros de la función cuadrática, se calcula de nuevo el corte o bisección
$c'_{0}$, como el valor mínimo de tal ecuación, haciendo:

\begin{equation}
\frac{d}{dx}\left(w_{0}x^{2}+y_{0}x+z_{0}\right)=0
\end{equation}

despejando $x$, y haciendo $c'_{0}=x$. Tal valor mínimo $f\left(c'_{0}\right)$ es guardado. Nuevamente se hace una bisección en el intervalo que contenga el punto mínimo, para repetir el proceso anterior. El proceso se detiene cuando el mínimo nuevo encontrado sea mayor al mínimo en la iteración anterior. La Figura muestra el caso donde la heurística de trazar una función cuadrática funciona correctamente donde el punto mínimo de la parabólica interpolada es el mejor de los puntos, pero otro en el que no, por lo que es necesario probar un intervalo distinto.

![](https://drive.google.com/uc?export=view&id=1QOeeXHBq6oUe6Oozcj77aoXgTeJV30Nh)

El algoritmo no garantiza su convergencia, pero si converge, lo hace relativamente rápido, sin necesidad de calcular gradientes para la
función a optimizar. 


```python
import math

def f(x):
    return x**2

a0 = -200.0
b0 = 25.0
siga = True
minimo = +math.inf
while siga:
    c0 = a0 + (b0-a0)/2
    fa0 = f(a0)
    fb0 = f(b0)
    fc0 = f(c0)

    print("a0: ", a0, "b0: ", b0, "c0: ", c0)
    print("fa0: ", fa0, "fb0: ", fb0, "fc0: ", fc0)

    # Sistema de ecuaciones
    # w0 * a0^2  +  y0 * a0  + z0  =  fa0
    # w0 * b0^2  +  y0 * b0  + z0  =  fb0
    # w0 * c0^2  +  y0 * c0  + z0  =  fc0
    a = np.array( [   [a0**2,a0,1],
                      [b0**2,b0,1],
                      [c0**2,c0,1]], dtype=float )
    b = np.array([fa0, fb0, fc0], dtype=float)
    x = np.linalg.solve(a,b)
    w0 = x[0]
    y0 = x[1]
    z0 = x[2]
    print("w0: ", w0, "y0: ", y0, "z0: ", z0)

    #  d      
    # --- ( w0 * x^2 + y0 * x + z0  )  =   0
    # dx

    # 2 * w0 * x  +  y0 = 0
    # x = -y0 / ( 2 * w0)
    x =  -y0 / ( 2* w0)
    c0 = x
    f1c0 = f(c0)
    print("c0:", c0)
    print("f1(c0): ", f1c0)
    if f1c0 < minimo:
        minimo = f1c0
        a0 = c0
    else:
        siga = False
print("Mínimo: ", f1c0)
    
```

## Optimización no convexa basada en la diferenciación: Algoritmo de descenso de gradiente

El algoritmo del descenso de gradiente sigue la idea de modificar el punto óptimo estimado de forma iterativa. Para una función en una
variable $f\left(x\right)$, la estimación del punto óptimo en una iteración $i+1$ está dada por: 

\begin{equation}
x\left(t+1\right)=x\left(t\right)-\alpha f'\left(x\left(t\right)\right)
\end{equation}

donde el coeficiente $\alpha$ determina el *grado de confianza o velocidad* con la que el proceso de optimización iterativa sigue
la dirección de la derivada. Para la optimización de una función multivariable $f\left(\overrightarrow{x}\left(t\right)\right)$ con $\overrightarrow{x}\in\mathbb{R}^{n}$, la posición óptima se estima usando el vector gradiente:

\begin{equation}
\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)-\alpha\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)
\end{equation}

La Figura muestra los pasos rectilíneos del algoritmo de descenso de gradiente, el cual sigue una dirección *recta*.

![](https://drive.google.com/uc?export=view&id=1lAwC9b5dZteIVPtjAWSzn5vEJUg1ERYk)

### Ejemplo

Tómese como ejemplo la función en ${\displaystyle \mathbb {R} ^{3}}$,

\begin{equation}
f\left(\overrightarrow{x}\right)=x_{1}^{2}-x_{2}^{2},
\end{equation}

la cual se grafica en la Figura siguiente


```python
def funcion(x, y):
  return x**2 - y**2


x = y = np.arange(-200.0, 200.0, 2.)
X, Y = np.meshgrid(x, y)
zs = np.array([funcion(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)


datos = []
trace = dict(
        name = "Nombre",
        x = X, y = Y, z = Z,
        type= 'surface'
)
datos.append( trace )
fig = go.Figure(data=datos)

fig.update_layout(

    scene=dict(
      xaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
      yaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
      zaxis=dict(
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)',
        showgrid = True
        ),
      aspectratio = dict( x=1, y=1, z=1 ) 
    )
)

fig.show()
```


Su gradiente se obtiene al hacer:

\begin{equation}
\nabla_{\overrightarrow{x}}f=\begin{bmatrix}2x_{1}\\
-2x_{2}
\end{bmatrix}
\end{equation}

Para un punto $\overrightarrow{x}\left(t\right)$ en la iteración $t=1$ elegido aleatoriamente, se actualiza la estimación del punto
óptimo de la siguiente forma:

\begin{equation}
\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)-\alpha\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)=\begin{bmatrix}x_{1}\left(t\right)\\
x_{2}\left(t\right)
\end{bmatrix}-\alpha\begin{bmatrix}2x_{1}\left(t\right)\\
-2x_{2}\left(t\right)
\end{bmatrix}=\begin{bmatrix}x_{1}\left(t\right)-\alpha2x_{1}\left(t\right)\\
x_{2}\left(t\right)+\alpha 2x_{2}\left(t\right)
\end{bmatrix}
\end{equation}

Suponiendo por ejemplo que $\alpha=0.25$ y que $\overrightarrow{x}\left(1\right)=\begin{bmatrix}200\\
0
\end{bmatrix}$, en tal caso $\overrightarrow{x}\left(2\right)=\begin{bmatrix}100\\
0
\end{bmatrix}$, por lo que son necesarias más iteraciones para converger en un punto
mínimo satisfactorio


**Punto silla**: Punto crítico donde una función tiene un máximo local en una dirección pero un mínimo local en la otra. El cual se señala en la Figura siguiente

![](https://drive.google.com/uc?id=1FKPjfsW91h0OT-uODF9TtI-ife6erbYv)

Fuente: Basado en https://es.wikipedia.org/wiki/Punto_de_silla

```python
# Ejemplo descenso de gradiente
# f =  ( x + 5 )**2
# código base: https://towardsdatascience.com/implement-gradient-descent-in-python-9b93ed7108d1

def derivada_f(x):
    return 2*(x+5)

cur_x = 3 # The algorithm starts at x=3
learning_rate = 0.01 # Learning rate
precision = 0.000001 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 10000 # maximum number of iterations
iters = 0 #iteration counter

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - learning_rate * derivada_f(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    if (iters % 40) == 0:
        print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)
```

## Variantes del algoritmo del descenso del gradiente: Descenso de gradiente adaptativo o AdaGrad y Propagación de la raíz media

Tómese la siguiente función multivariable con dominio $\overrightarrow{x}\in\mathbb{R}^{2}$,

\begin{equation}
f\left(\overrightarrow{x}\right)=0.1x_{1}^{2}+2x_{2}^{2},
\end{equation}

graficada en la Figura

![](https://drive.google.com/uc?export=view&id=1eOnONLJIyjbAXluo32NifKzLToyKL9a1)

Al calcular el vector gradiente $\nabla_{\overrightarrow{x}}f=\begin{bmatrix}0.2x_{1}\\
4x_{2}
\end{bmatrix}$, podemos notar que el desplazamiento será siempre más fuerte hacia la dimensión $x_{2}$. Esto ocasionará *saltos* más fuertes en tal dimensión, sobre todo con coeficientes de actualización $\alpha$ altos.


### AdaGrad

Existen distintos enfoques para para solucionar el problema de distintas magnitudes de gradiente a distintas direcciones, y sus consecuentes oscilaciones en el espacio de búsqueda. Por ejemplo, *Duchi et al. (2011). "Adaptive subgradient methods for online learning and stochastic optimization"*. La técnica conocida popularmente como AdaGrad, propone utilizar un coeficiente de aprendizaje particular por cada dimensión 
  
\begin{equation}
\overrightarrow{\alpha}=\begin{bmatrix}\alpha_{1}\\
\vdots\\
\alpha_{n}
\end{bmatrix},
\end{equation}
  
\begin{equation}
\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)-\overrightarrow{\alpha}.*\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right),
\end{equation}
  
donde el operador $.*$ multiplica los vectores por elemento (element-wise), a diferencia del algoritmo del descenso de gradiente original, que utiliza un sólo coeficiente de aprendizaje escalar, lo que implica el uso del mismo valor en todas las dimensiones $\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)-\alpha\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)$.
Los elementos del vector $\overrightarrow{\alpha}$, AdaGrad propone calcularlos adaptativamente, normalizando el coeficiente de aprendizaje para la dimensión $i$, por la sumatoria de todas las derivadas parciales evaluadas en cada punto $\overrightarrow{x}\left(t\right)$ hasta la iteración actual $t$: 
  
\begin{equation}
\alpha_{i}=\frac{\rho}{\sqrt{s_{i}\left(t\right)+\epsilon}}
\end{equation}
  
con un valor pequeño $\epsilon$ para evitar las divisiones por cero (estabilidad numérica). El término $s_{i}\left(\tau\right)$ es el
que contiene las magnitudes de las derivadas parciales en la solución candidata actual:
  
\begin{equation}
s_{i}\left(t\right)=\sum_{k=1}^{t}\left(\frac{df\left(\overrightarrow{x}\left(k\right)\right)}{dx_{i}}\right)^{2}
\end{equation}
  
El coeficiente $\rho$ controla el aprendizaje global del algoritmo. De esta forma, si un componente del gradiente tiene una alta magnitud, esto hará que su $\alpha_{i}$ sea más pequeño. Una desventaja del método AdaGrad es que cada coeficiente $\alpha_{i}$ tiende a hacerse más chico conforme pasan más iteraciones, al dar igual peso a los valores de la derivada parcial anteriores.


### RMS Prop

El algoritmo **Root Mean Square Propagation (RMS Prop)**, introducido en "Tieleman, et al. (2012). Lecture 6.5-rmsprop: Divide the gradient by a running average of its recent magnitude, coursera" propone entonces usar el pesado exponencial móvil con lo que entonces, se calcula el término *$s_{i}\left(\tau\right)$* de la siguiente forma:
  
\begin{equation}
s_{i}\left(t\right)=\gamma s_{i}\left(t-1\right)+\left(1-\gamma\right)\sum_{t=1}^{\tau}\left(\frac{df\left(\overrightarrow{x}\left(k\right)\right)}{dx_{i}}\right)^{2}
\end{equation}
  

## Optimización no convexa basada en la diferenciación: Algoritmo de Newton-Raphson
Antes de discutir el método de Newton-Raphson, es necesario repasar el concepto de series, específicamente, las series de Taylor.

### Series de Taylor

Para una función $f\left(x\right)$, las series de Taylor permiten aproximar a una función alrededor de un punto de operación $a$ de
la siguiente forma: 

\begin{equation}
f\left(x\right)=\sum_{k=0}^{\infty}\frac{f^{(k)}\left(a\right)\left(x-a\right)^{k}}{k!}=f\left(a\right)+f'\left(a\right)\left(x-a\right)+\frac{f''\left(a\right)}{2!}\left(x-a\right)^{2}+\ldots
\end{equation}

Con lo que entonces la aproximación de primer orden o linear alrededor del punto $a$ toma sólo dos términos: 

\begin{equation}
f\left(x\right)\approx f\left(a\right)+f'\left(a\right)\left(x-a\right).
\end{equation}

Entre más alejados estén los valores de $x$ respecto al punto $a$, mayor será la diferencia entre la función real y la aproximación
por series de Taylor. Para una función multivariable con $\vec{x}=\left[x_{1},x_{2},\ldots,x_{n}\right]$
la aproximación de Taylor de primer orden alrededor de un punto $\overrightarrow{a}=\left[a_{1},a_{2},\ldots,a_{n}\right]^{T}$
está dada por:

\begin{equation}
f\left(\vec{x}\right)=f\left(\overrightarrow{a}\right)+\frac{\partial f}{\partial x_{1}}\left(x_{1}-a_{1}\right)+\frac{\partial f}{\partial x_{2}}\left(x_{2}-a_{2}\right)+\ldots+\frac{\partial f}{\partial x_{n}}\left(x_{n}-a_{n}\right).
\end{equation}



### El método de Newton-Raphson

El método de Newton-Raphson fue diseñado originalmente para encontrar la raíz de ecuaciones no lineales igualadas a cero: 

\begin{equation}
f\left(x\right)=0
\end{equation}

y para el caso de una función multivariable, $f\left(\overrightarrow{x}\right)=0$. Su extensión para el problema de encontrar un óptimo en tal función, viene dado al usar el algoritmo para encontrar la raíz de la primera derivada, resolviendo entonces

\begin{equation}
f'\left(\overrightarrow{x}\right)=0
\end{equation}

El algoritmo de Newton-Raphson es iterativo, de forma que a partir de un punto inicial $x\left(1\right)$, calcula una secuencia de valores $x\left(1\right),x\left(2\right),x\left(3\right),\ldots,x\left(p\right)$, con $x\left(p\right)\rightarrow x_{\textrm{opt}}$. Por cada iteración $i$, el punto óptimo estimado se calcula como sigue: 

\begin{equation}
x\left(t+1\right)=x\left(t\right)+\Delta x
\end{equation}

Para obtener el valor de $\Delta x$ se descompone la función $f\left(x\right)$ utilizando una aproximación por series de Taylor, para hallar el $\Delta x$ óptimo igualando tal aproximación de Taylor para igualarla a cero. Recordemos que la serie de Taylor está dada en general por:

\begin{equation}
f\left(x\right)=\sum_{k=0}^{\infty}\frac{f^{(k)}\left(a\right)}{k!}\left(x-a\right)^{k}=f\left(a\right)+f'\left(a\right)\left(x-a\right)+\frac{f''\left(a\right)}{2!}\left(x-a\right)^{2}+\ldots
\end{equation}

en este caso de segundo grado, con $a=x\left(t\right)$ como punto de operación, sobre el intervalo $\left[x_{i},x_{i}+\Delta x\right]$:

\begin{equation}
f\left(x\left(t\right)\right)\approx f\left(x\left(t\right)\right)+f'\left(x\left(t\right)\right)\left(\left(x\left(t\right)+\Delta x\right)-x\left(t\right)\right)+\frac{f''\left(x\left(t\right)\right)}{2}\left(\left(x\left(t\right)+\Delta x\right)-x\left(t\right)\right)^{2}=f\left(x\left(t\right)\right)+f'\left(x\left(t\right)\right)\Delta x+\frac{f''\left(x\left(t\right)\right)}{2}\Delta x^{2}
\end{equation}

Observe que esta aproximación es sólo válida en tal intervalo. Derivando respecto al $\Delta x$ e igualando a cero, en búsqueda del punto
óptimo, se tiene que:

\begin{equation}
\frac{d}{d\Delta x}\left(f\left(x\left(t\right)\right)+f'\left(x\left(t\right)\right)\left(\Delta x\right)+\frac{f''\left(x\left(t\right)\right)}{2}\Delta x^{2}\right)=0
\end{equation}

\begin{equation}
\Rightarrow f'\left(x\left(t\right)\right)+f''\left(x\left(t\right)\right)\Delta x=0\Rightarrow\Delta x=-\frac{f'\left(x\left(t\right)\right)}{f''\left(x\left(t\right)\right)}
\end{equation}

Al realizar este cálculo, hallaremos el $\Delta x$ que se acerque más a la solución de $f'\left(x\right)=0$. Para el caso
de la optimización de una función multivariable $f\left(\overrightarrow{x}\left(t\right)\right)$ con $\overrightarrow{x}\left(t\right)\in\mathbb{R}^{n\times1}$, el cambio en el vector $\overrightarrow{x}_{i}$ viene dado en términos de la matriz Hessiana $H_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)\in\mathbb{R}^{n\times n}$
evaluada en tal punto $\overrightarrow{x}\left(t\right)$:

\begin{equation}
\Delta\overrightarrow{x}=-\left[H_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)\right]^{-1}\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)
\end{equation}

donde el vector gradiente tiene dimensiones $\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)\in\mathbb{R}^{n\times1}$,
lo cual hace que $\Delta\overrightarrow{x}\in\mathbb{R}^{n\times1}$. La ecuación de actualización del vector viene entonces dada por: 
\begin{equation}
\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)+\Delta\overrightarrow{x}
\end{equation}


### Ejemplo

Tómese una función con dominio $\overrightarrow{x}\in\mathbb{R}^{2}$,

\begin{equation}
f\left(\overrightarrow{x}\right)=x_{1}^{2}-x_{2}^{2},
\end{equation}


Formalmente, para encontrar el punto mínimo en una función usando el método de Newton-Raphson, se utiliza la aproximación por series
de Taylor del gradiente para hacer $g=\nabla_{\overrightarrow{x}}\left(f\right)=0$. Sin embargo, esto involucraría derivar dos veces más la función $g$, la cuál ya es vectorial, con lo que se complica el cómputo. Es por ello que nos quedaremos con la búsqueda con el punto $f\left(\overrightarrow{x}\right)=0$, lo cuál nos permite estimar el $\Delta x$ haciendo: 

\begin{equation}
\Delta\overrightarrow{x}=-\left[H_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)\right]^{-1}\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)
\end{equation}

para por cada iteración actualizar el punto óptimo estimado haciendo
$\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)+\Delta\overrightarrow{x}$.

Calculando ambos factores se obtiene: 

\begin{equation}
\nabla_{\overrightarrow{x}}f=\begin{bmatrix}2x_{1}\\
-2x_{2}
\end{bmatrix}
\end{equation}

y la matriz Hessiana está dada por: 

\begin{equation}
H_{\overrightarrow{x}}=\begin{bmatrix}\frac{df}{dx_{1}^{2}} & \frac{df}{dx_{1}dx_{2}}\\
\frac{df}{dx_{2}dx_{1}} & \frac{df}{dx_{2}^{2}}
\end{bmatrix}=\begin{bmatrix}2 & 0\\
0 & -2
\end{bmatrix}
\end{equation}

 y su inversa por $H_{\overrightarrow{x}}^{-1}=\begin{bmatrix}\frac{1}{2} & 0\\
0 & -\frac{1}{2}
\end{bmatrix}$. Para un punto cualquiera en la iteración $t=1$, se tiene que:

\begin{equation}
\Delta\overrightarrow{x}=-\left[H_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)\right]^{-1}\nabla_{\overrightarrow{x}}f\left(\overrightarrow{x}\left(t\right)\right)=-\begin{bmatrix}\frac{1}{2} & 0\\
0 & -\frac{1}{2}
\end{bmatrix}\begin{bmatrix}2x_{1}\left(t\right)\\
-2x_{2}\left(t\right)
\end{bmatrix}=\begin{bmatrix}-x_{1}\left(t\right)\\
-x_{2}\left(t\right)
\end{bmatrix}=-\overrightarrow{x}\left(t\right)
\end{equation}
y al hacer la actualización

\begin{equation}
\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)+\Delta\overrightarrow{x}\Rightarrow\overrightarrow{x}\left(t+1\right)=\overrightarrow{x}\left(t\right)-\overrightarrow{x}\left(t\right)=\begin{bmatrix}0\\
0
\end{bmatrix}
\end{equation}

Lo cual significa que el algoritmo ha arribado al punto silla, el cual en este caso no es el óptimo, pero lo hizo en una iteración
sin importar el punto inicial $\overrightarrow{x}\left(t\right)$. Las funciones de error usualmente no tienen partes negativas, pero
si muchos puntos silla.
